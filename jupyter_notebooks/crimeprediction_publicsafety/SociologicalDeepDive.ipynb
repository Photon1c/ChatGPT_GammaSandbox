```python
import requests
import pandas as pd
import io

# Specify the API endpoint and query parameters
endpoint = "https://wonder.cdc.gov/"
params = {
    "stage": "request",
    "action": "execute",
    "format": "csv",
    "requested_tables": ["VSUICIDE"],
    "query": [
        {
            "code": "YEAR",
            "selection": {
                "filter": "item",
                "values": [str(year) for year in range(2010, 2021)]
            }
        },
        {
            "code": "AGE",
            "selection": {
                "filter": "item",
                "values": ["-0", "1-4", "5-11", "12-14", "15-19", "20-24", "25-34", "35-44", "45-54", "55-64", "65-74", "75+"]
            }
        },
        {
            "code": "SEX",
            "selection": {
                "filter": "item",
                "values": ["MF"]
            }
        },
        {
            "code": "RACE",
            "selection": {
                "filter": "item",
                "values": ["-1", "1", "2", "3", "4", "5", "6", "7", "A", "B", "D", "O"]
            }
        }
    ]
}

# Send the request and retrieve the response as a DataFrame
response = requests.get(endpoint, params=params)
content = response.content.decode("ISO-8859-1")
skiprows = content.find("\n") + 1 if "\n" in content else 0
df = pd.read_csv(io.StringIO(content), skiprows=skiprows, error_bad_lines=False, warn_bad_lines=True)

# Write the data to a CSV file
df.to_csv("suicide_statistics.csv", index=False)
```
