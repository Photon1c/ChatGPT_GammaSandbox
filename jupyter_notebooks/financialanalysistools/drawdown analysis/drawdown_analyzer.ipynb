```python
#Calculate drawdown for given values
#From https://en.wikipedia.org/wiki/Drawdown_(economics)
import matplotlib.pyplot as plt

def calculate_drawdown(nav):
    n = len(nav)
    dd = [0] * n  # Initialize Drawdown list
    mdd = 0
    peak = -99999

    for i in range(n):
        # Update peak if the current NAV is higher
        if nav[i] > peak:
            peak = nav[i]

        # Calculate Drawdown for the current time step
        dd[i] = 100.0 * (peak - nav[i]) / peak

        # Update Max Drawdown if the current Drawdown is higher
        if dd[i] > mdd:
            mdd = dd[i]

    return dd, mdd

# Example usage:
nav_values = [100, 110, 90, 120, 80, 130]
drawdown_values, max_drawdown = calculate_drawdown(nav_values)

# Plot Net Asset Value
plt.plot(nav_values, label='Net Asset Value', marker='o')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Net Asset Value')
plt.legend()
plt.title('Net Asset Value Over Time')

# Show the plot
plt.show()


print("Drawdown values:", drawdown_values)
print("Max Drawdown:", max_drawdown)
```
