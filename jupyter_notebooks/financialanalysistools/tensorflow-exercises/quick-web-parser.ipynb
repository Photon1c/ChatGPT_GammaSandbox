```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#create list of web element in page

def get_nested_elements(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    elements = soup.find_all()
    return elements

def categorize_elements(elements):
    categories = {}
    for element in elements:
        category = element.name
        if category not in categories:
            categories[category] = []
        categories[category].append(str(element))
    return categories

if __name__ == '__main__':
    url = 'https://Google.com'
    elements = get_nested_elements(url)
    categories = categorize_elements(elements)
    df = pd.DataFrame([(category, len(categories[category])) for category in categories],
                      columns=['Category', 'Count'])
    print(df)
```
