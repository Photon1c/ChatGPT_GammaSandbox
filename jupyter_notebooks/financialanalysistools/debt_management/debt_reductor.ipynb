```python
import matplotlib.pyplot as plt 
import numpy as np

def debtpay(initial_debt, interest, monthly_payment, number_of_payments):
    debt = []
    debt.append(initial_debt) 
    for i in range(1, number_of_payments + 1):
        interest_amount = initial_debt * (interest / 12 / 100)
        to_be_paid = min(initial_debt, monthly_payment)
        initial_debt = initial_debt - to_be_paid + interest_amount
        debt.append(initial_debt)
    return debt  # add this line to return the debt list

initial_debt = 100000          
interest = 25/100            
monthly_payment = 3000        
number_of_payments = 60     

debt = debtpay(initial_debt, interest, monthly_payment, number_of_payments)  # store the returned debt list in a variable

plt.plot(range(number_of_payments + 1), debt)  # add 1 to the range to include the initial debt
plt.xlabel("Month")  
plt.ylabel("Debt")

plt.show()
```
