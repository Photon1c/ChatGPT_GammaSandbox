```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load the input data from a CSV file
input_data = pd.read_csv('input_data.csv')

# extract the input features and output variable
input_features = input_data[['Age', 'Income']].values
output_variable = input_data['Medical Condition'].values.reshape(-1, 1)

# define the model architecture
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, input_dim=2, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# compile the model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model on the input and output data
model.fit(input_features, output_variable, epochs=100)

# load new data to test the model
new_data = pd.read_csv('new_data.csv')
new_input_features = new_data[['Age', 'Income']].values

# make predictions on the new data
predictions = model.predict(new_input_features)

# print the prediction results in a user-friendly format
for i in range(len(predictions)):
    if predictions[i] >= 0.5:
        print("Person {} is likely to have the medical condition with a probability of {:.2f}%".format(i+1, predictions[i][0]*100))
    else:
        print("Person {} is unlikely to have the medical condition with a probability of {:.2f}%".format(i+1, (1-predictions[i][0])*100))

# plot a bar chart of the predicted probabilities
probabilities = [predictions[i][0] for i in range(len(predictions))]
plt.bar(range(len(probabilities)), probabilities)
plt.xticks(range(len(probabilities)), ['Person 1', 'Person 2'])
plt.ylim([0, 1])
plt.ylabel('Probability')
plt.title('Predicted Probabilities for Medical Condition')
plt.show()
```
