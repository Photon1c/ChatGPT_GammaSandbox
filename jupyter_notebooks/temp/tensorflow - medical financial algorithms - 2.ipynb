```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# define the input data as a numpy array
input_data = np.array([[40, 50000], [35, 75000], [50, 100000], [45, 60000], [30, 80000]])

# define the output data as a numpy array
output_data = np.array([[1], [0], [1], [0], [0]])

# define the model architecture
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, input_dim=2, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# compile the model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model on the input and output data
model.fit(input_data, output_data, epochs=100)

# test the model on some new data
new_data = np.array([[42, 90000], [25, 40000]])
predictions = model.predict(new_data)

# print the prediction results in a user-friendly format
for i in range(len(predictions)):
    if predictions[i] >= 0.5:
        print("Person {} is likely to have the medical condition with a probability of {:.2f}%".format(i+1, predictions[i][0]*100))
    else:
        print("Person {} is unlikely to have the medical condition with a probability of {:.2f}%".format(i+1, (1-predictions[i][0])*100))

# plot a bar chart of the predicted probabilities
probabilities = [predictions[i][0] for i in range(len(predictions))]
plt.scatter(range(len(probabilities)), probabilities)
plt.xticks(range(len(probabilities)), ['Person 1', 'Person 2'])
plt.ylim([0, 1])
plt.ylabel('Probability')
plt.title('Predicted Probabilities for Medical Condition')
plt.show()
```
